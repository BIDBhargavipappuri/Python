```python
import requests

API_KEY = "c728e5afd9e383b3ba542f7d409ac70a"
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

# Ask user for multiple cities
cities_input = input("Enter city names separated by commas: ")
cities = [city.strip() for city in cities_input.split(",")]

weather_data = {}   # store results for comparison

for city in cities:
    params = {
        "q": city,
        "appid": API_KEY,
        "units": "metric"
    }

    response = requests.get(BASE_URL, params=params)
    data = response.json()

    print(f"\n--- Weather Report for {city} ---")

    if response.status_code == 200:
        temperature = data["main"]["temp"]
        humidity = data["main"]["humidity"]
        wind_speed = data["wind"]["speed"]
        condition = data["weather"][0]["description"]

        print(f"Temperature is: {temperature} °C")
        print(f"Humidity is: {humidity}%")
        print(f"Wind speed is: {wind_speed} m/s")
        print(f"Conditions are: {condition}")

        # Store for comparison
        weather_data[city] = {
            "temp": temperature,
            "humidity": humidity,
            "wind": wind_speed,
            "condition": condition
        }

    else:
        print(f"Error: {data.get('message', 'Unable to fetch weather')}")

# -------------------------------
# COMPARISON SECTION
# -------------------------------

if len(weather_data) > 1:
    print("\n==============================")
    print("      WEATHER COMPARISON      ")
    print("==============================")

    # Compare temperature
    hottest_city = max(weather_data, key=lambda c: weather_data[c]["temp"])
    coldest_city = min(weather_data, key=lambda c: weather_data[c]["temp"])

    print(f"\nHottest city: {hottest_city} ({weather_data[hottest_city]['temp']} °C)")
    print(f"Coldest city: {coldest_city} ({weather_data[coldest_city]['temp']} °C)")

    # Compare humidity
    most_humid = max(weather_data, key=lambda c: weather_data[c]["humidity"])
    least_humid = min(weather_data, key=lambda c: weather_data[c]["humidity"])

    print(f"\nMost humid city: {most_humid} ({weather_data[most_humid]['humidity']}%)")
    print(f"Least humid city: {least_humid} ({weather_data[least_humid]['humidity']}%)")

    # Compare wind speed
    windiest = max(weather_data, key=lambda c: weather_data[c]["wind"])
    calmest = min(weather_data, key=lambda c: weather_data[c]["wind"])

    print(f"\nWindiest city: {windiest} ({weather_data[windiest]['wind']} m/s)")
    print(f"Calmest city: {calmest} ({weather_data[calmest]['wind']} m/s)")
```
