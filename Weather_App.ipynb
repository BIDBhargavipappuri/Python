```python
import requests

API_KEY = "c728e5afd9e383b3ba542f7d409ac70a"
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

# Ask user for city name
city = input("Enter city name: ")

params = {
    "q": city,
    "appid": API_KEY,
    "units": "metric"   # Celsius
}

response = requests.get(BASE_URL, params=params)
data = response.json()

if response.status_code == 200:
    temperature = data["main"]["temp"]
    humidity = data["main"]["humidity"]
    wind_speed = data["wind"]["speed"]
    condition = data["weather"][0]["description"]

    print("\n--- Weather Report ---")
    print(f"City: {city}")
    print(f"Temperature is: {temperature} °C")
    print(f"Humidity is: {humidity}%")
    print(f"Wind speed is: {wind_speed} m/s")
    print(f"Conditions are: {condition}")

else:
    # Handling errors when status code != 200
    print("\n--- Error Fetching Weather ---")
    print(f"Status Code: {response.status_code}")
    print(f"Reason: {data.get('message', 'Unknown error occurred')}")
    print("Please check the city name or try again later.")
```
